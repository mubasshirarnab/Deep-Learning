```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datapath = '/content/drive/MyDrive/Colab Notebooks/perceptronDataset.csv'
```

```python
dataFrame = pd.read_csv(datapath);
print(dataFrame)
```

```python
dataSet = dataFrame.to_numpy()
X = dataSet[:, 0:4]
y = dataSet[:, 4]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
```

```python
def perceptron(X,y):
  X = np.insert(X, 0, 1, axis=1)
  w = np.ones(X.shape[1])
  learning_rate = 0.1

  for i in range(1000):
    a = np.random.randint(0, X.shape[0])
    y_pred = step(np.dot(X[a], w))
    w = w + learning_rate * (y[a] - y_pred) * X[a]
  return w[0],w[1:]
```

```python
def step(value):
  return 1 if value > 0 else 0
```

```python
intercept, coefficients = perceptron(X, y)
print(intercept, coefficients)
```

```python
m = -(coefficients[0] / coefficients[1])
c = -(intercept / coefficients[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m * x_input + c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
